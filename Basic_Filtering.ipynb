```python
import numpy as np
```

Task 1.1: Basic Filtering

```python
ages = np.array([25, 30, 18, 35, 22, 40, 17])
ages
```

```python
adutls  = np.array([ages > 30])
adutls
```

Task 1.2: Filtering with Multiple Conditions

```python
young_adults = ages[(ages > 20) & (ages < 35)]
young_adults
```

Task 1.3: Filtering Strings

```python
name = np.array(['Alice', 'Bob', 'Charlie', 'David', 'Eve'])
name
```

```python
n = (np.char.startswith(name,'A')) | (np.char.startswith(name,'E'))
name[n]
```

Task 2.1: Basic Masking for Selection

```python
temperatures = np.array([28, 32, 25, 35, 30, 22, 38])
temperatures
```

```python
is_not = temperatures >30
is_not
```

```python
hot_Temperature = temperatures[is_not]
hot_Temperature
```

Task 2.2: Masking for Modification

```python
for i in range(len(hot_Temperature)):
  hot_Temperature[i]=31
```

```python
print("Change Celsius 30 Degree above is set  31 Celsius ",hot_Temperature)
```

Task 2.3: Using np.where() for Masking

```python
scores = np.array([65, 80, 72, 90, 55, 78])
scores
```

```python
grades = np.where(scores > 75,"Pass","Fail")
grades
```

Task 4.2: Broadcasting a 1D Array to a 2D Array (Row-wise)

```python
sales = np.array([[10,20],[15,25],[12,18]])
sales
```

```python
sales.shape
```

```python
bonuses = np.array([[[1],[2],[3]],[[1],[1],[1]]])
bonuses
```

```python
bonuses.shape
```

```python
add_bonuses = sales + bonuses
add_bonuses
```

```python
add_tax = sales * tax_rates
add_tax
```
