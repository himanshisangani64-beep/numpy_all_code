5)  Implement a function using NumPy that calculates the Euclidean distance between two points.

```python

def Euclidean_Distance(l,l1):
  return np.sqrt(np.sum((l-l1)**2))


l = np.array([1,6,12])
l1 = np.array([2,4,8])

e1 = Euclidean_Distance(l1,l)
e1
```
